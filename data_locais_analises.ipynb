```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
columns= ['numero_boletim','data_boletim','numero_municipio','nome_municipio','sequencia_logradouros','numero_logradouro','tipo_logradouro','nome_logradouro','tipo_logradouro_anterior','nome_logradoro_anterior','Numero_bairro','nome_bairro','tipo_bairro','descricao_tipo_bairro','numero_imovel','numero_imovel_proximo']
```

```python
df_locais_2011 = pd.read_csv('./dataset_logradouros_acidentes/si-log-2011.csv', names=columns, usecols=columns,
parse_dates=['data_boletim'],skiprows=1, encoding = "ISO 8859-1", sep=';')
df_locais_2012 = pd.read_csv('./dataset_logradouros_acidentes/si-log-2012.csv', names=columns, usecols=columns,
parse_dates=['data_boletim'],skiprows=1, encoding = "ISO 8859-1", sep=';')
df_locais_2013 = pd.read_csv('./dataset_logradouros_acidentes/si-log-2013.csv', names=columns, usecols=columns,
parse_dates=['data_boletim'],skiprows=1, encoding = "ISO 8859-1", sep=';')
df_locais_2014 = pd.read_csv('./dataset_logradouros_acidentes/si-log-2014.csv', names=columns, usecols=columns,
parse_dates=['data_boletim'],skiprows=1, encoding = "ISO 8859-1", sep=';')
df_locais_2015 = pd.read_csv('./dataset_logradouros_acidentes/si-log-2015.csv', names=columns, usecols=columns,
parse_dates=['data_boletim'],skiprows=1, encoding = "ISO 8859-1", sep=';')
```

```python
df_locais_final = pd.concat([df_locais_2011,df_locais_2012,df_locais_2013,df_locais_2014,df_locais_2015])
df_locais_final.set_index(['numero_boletim'])
```

```python
df_locais_final.info()
```

```python
df_locais_final = df_locais_final.applymap(lambda x: x.strip() if type(x) is str else x)
```

```python
df_locais_final.dropna(inplace=True)
del df_locais_final['tipo_bairro']
del df_locais_final['nome_logradoro_anterior']
del df_locais_final['tipo_logradouro_anterior']
```

```python
df_locais_final.to_csv('./dataset_logradouros_acidentes/data_logradouro.csv')
```
